```python
from bs4 import BeautifulSoup

from sbdetection.models import baseline
```

```python
from typing import Union, Callable

def generate_html(html: Union[str, bytes], predictor: Callable):
    soup = BeautifulSoup(html)
    for p in soup.select('p'):
        sentences = predictor(p.text)
           
        empty_paragraph = BeautifulSoup('<p></p>')
        paragraph = empty_paragraph.p
        for sentence in sentences:
            new_tag = soup.new_tag("span")
            new_tag.string = sentence
            paragraph.append(new_tag)

        p.replace_with(paragraph)
        
    return soup
```

```python
test = """
<html>
    <header> this is a test</header>
    <body>
        <p> Hello world! I need part 2 also.</p>
    </body> 
</html>
"""

generate_html(test, baseline.predict_sentences)
```

```python
import requests

result = requests.get("http://oreilly.com/store/samplers.html")
c = result.content

generate_html(c, baseline.predict_sentences)
```
